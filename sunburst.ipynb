```python
import pandas as pd
import plotly.express as px
%config InlineBackend.figure_format = 'retina'
```

```python
accounts_df = pd.read_csv("data/accounts.csv")
plt_df = accounts_df.groupby(['vertical', 'subvertical']).size().reset_index().rename(columns={0:'count'})
```

```python
# Proportion
fig = px.sunburst(plt_df, path=['vertical', 'subvertical'], 
                  values='count')
fig.update_layout(
    autosize=False,
    width=1000,
    height=1000
)
fig.show();
```
