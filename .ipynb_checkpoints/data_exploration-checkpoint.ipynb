## Objective

* Explore accounts table - `gc-prd-ct-data-share-prod-34fe.core_model.d_salesforce_accounts`
* Explore opportunities table - `gc-prd-ct-data-share-prod-34fe.core_model.d_salesforce_opportunities`


### Filters

We want to make certain inclusions and exclusions in our training pool. 

* **Inbound** opportunities only - filtering on the *source_bucket* column
* **New** opportunities only - filtering on the *opportunity_type* column
* **VRA** opportunities only - filtering on the *commission_model* column



### Target

We will focus for now on just the first two, ‘closed won’ and ‘closed lost’ from the opportunity_stage_name column, filtering out the rest. Of the remaining, 42% are ‘closed won’.

```python
def agg_plot_weekly(x):
    x = pd.to_datetime(x)
    x = x.dt.date
    x = x.value_counts()
    x.index = pd.to_datetime(x.index)
    x.resample('W').sum().plot();
```

## Target variable

```python
accounts_df['closed_won'] = np.where(accounts_df['opportunity_stage_name'] == '7. Closed Won', 1, 0)
```

```python
# Overall mean
accounts_df['closed_won'].mean()
```

```python
# Check opportunities table
opportunities_df['closed_won'] = np.where(opportunities_df['opportunity_stage_name'] == '7. Closed Won', 1, 0)
```

```python
opportunities_df['closed_won'].mean()
```

```python
# Convert date
accounts_df['opportunity_created_date'] = [re.sub(' UTC', '', i) for i in accounts_df['opportunity_created_date']]
accounts_df['opportunity_created_date'] = pd.to_datetime(accounts_df['opportunity_created_date']).dt.date
accounts_df['MONTH_YEAR'] = accounts_df['opportunity_created_date'].astype('datetime64[M]')
```

```python
# Get counts by month
counts_by_month = accounts_df['opportunity_created_date'].value_counts()
counts_by_month.index = pd.to_datetime(counts_by_month.index)
counts_by_month = counts_by_month.resample('M').sum()
```

```python
# plot
fig, ax = plt.subplots(figsize=(10,5))
counts_by_month.plot(ax=ax)
ax.set_ylabel("Number of opportunities")
ax.set_xlabel("Month")
plt.title("Number of opportunities by month")
plt.show();
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10,5))
accounts_df.groupby('MONTH_YEAR')['closed_won'].mean().resample("M").sum().plot(ax=ax, color="#272635")
ax.set_ylabel("Conversion Rate")
ax.set_xlabel("Mo`nth")
plt.title("Conversion rate by month")
plt.show();
```
